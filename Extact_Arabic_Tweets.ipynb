
<h1>EXTRACT ARABIC SPORTIF TWEETS</h1>

```python
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
import re
import json

# Enter Twitter API Keys
consumer_key='confidential'
consumer_secret='confidential'
access_token='confidential'
access_secret='confidential'

# Create tracklist with the words that will be searched for
tracklist = ['Ø§Ù„Ø±ÙŠØ§Ø¶Ø©','kooora','ÙƒØ±Ø© Ø§Ù„Ù‚Ø¯Ù…','beinsport','soccer','sport','Ø§Ù„ÙÙŠÙØ§']
# Initialize Global variable
tweet_count = 0
# Input number of tweets to be downloaded
n_tweets =500

# Create the class that will handle the tweet stream
class StdOutListener(StreamListener):
      
    def on_data(self, data):
        global tweet_count
        global n_tweets
        global stream
        all_data=[]
        
             
        if tweet_count < n_tweets:
            #print(data)
            #all_data = json.loads(data)
            with open('arabic_sportif_tweets.json', 'a') as f:
                f.write(data)
           
            
             
            
            
            tweet_count += 1
            
        
   
            
            return True
        else:
            stream.disconnect()
    

    def on_error(self, status):
        print(status)



# Handles Twitter authetification and the connection to Twitter Streaming API
l = StdOutListener()
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
stream = Stream(auth, l)

stream.filter(track=tracklist,languages=["ar"])
```
